```python
print("my netflix project")
import pandas as pd
import matplotlib.pyplot as plt
df =pd.read_csv("netflix_titles.csv")
print("data loaded")
print("rows:" ,df.shape[0])
print("columns:" ,df.shape[1])
df.head()
df.columns
df['type'].value_counts()
df['type'].value_counts().plot(kind='bar', color=['#E50914', '#221F1F'])
plt.title ('Movies vs tv shows on netflix')
plt.xlabel('type')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.show()
df['country'].value_counts().head(10)
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 countries on Netflix')
plt.xlabel('country')
plt.ylabel('count')
plt.xticks(rotation =45)
plt.show()
df['rating'].value_counts().plot(kind='bar')
plt.title('Movies vs tv shows Ratings' )
plt.xlabel('ratings')
plt.ylabel('titles')
plt.xticks(rotation=45)
plt.show()
recent = df[df['release_year']>=2010]
recent['release_year'].value_counts().sort_index().plot(kind='bar')
plt.title('Netflix Content Growth by Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("Key Insights from Netflix Data Analysis")
print("1. Netflix has 6131 Movies vs 2676 TV Shows")
print("2. USA is the top content producer, followed by India")
print("3. TV-MA is the most common rating — Netflix targets adults")
print("4. Content additions peaked in 2018")
```
